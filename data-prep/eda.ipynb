```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Setting
sns.despine(left=True, bottom=True)
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format','{:.4f}'.format)
sns.set(rc={"figure.dpi":300, 'savefig.dpi':300})
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
sns.set()
```

```python
customers = pd.read_parquet('.\\tables\\customers.parquet')
employee =  pd.read_parquet('.\\tables\\employee.parquet')
order_detail = pd.read_parquet('.\\tables\\order_detail.parquet')
orders = pd.read_parquet('.\\tables\\orders.parquet')
shippers = pd.read_parquet('.\\tables\\shippers.parquet')
```

```python
order_detail['sales'] = order_detail.UnitPrice * order_detail.Quantity
order_detail['sales_after_disc'] = order_detail.sales * (1 - order_detail.Discount)
```

```python
orders = pd.merge(orders, shippers[['ShipperID', 'CompanyName']], left_on='ShipVia', right_on='ShipperID')
```

```python
orders = orders.rename(columns={'CompanyName': 'shipper'})
orders = orders.drop(['ShipVia', 'ShipperID'], axis=1)
orders['is_late'] = orders.RequiredDate < orders.ShippedDate
```

```python
employee.head(1)
```

```python
df_all = pd.merge(orders, order_detail, on='OrderID', how='right')
df_all = pd.merge(df_all, customers, on='CustomerID', how='left')
df_all = pd.merge(df_all, employee, on='EmployeeID', how='left')
```

```python
df_all = df_all.reset_index(drop=True)
df_all.to_csv('.\\tables\\df_all.csv', index=False)
```
